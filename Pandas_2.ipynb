**Apply function:The apply() method is one of the most common methods of data preprocessing. It simplifies applying a function on each element in a pandas Series and each row or column in a pandas DataFrame.**

```python
import pandas as pd
import numpy as np


students = pd.Series(data=[180, 175, 168, 190],
                     index=['Vik', 'Mehdi', 'Bella', 'Chriss'])
display(students)
print(type(students))
```

```python
def cm_to_feet(h):
    return np.round(h/30.48, 2)

print(students.apply(cm_to_feet))
```

**What is Pandas map()?
pandas.series.map maps values of Series according to an input mapping function. Used for substituting each value in a Series with another value, that may be derived from a function, a dict or a Series.**

```python
#
import pandas as pd
df = pd.DataFrame({'name':['John Doe', 'Mary Re', 'Harley Me'],
                   'gender':[1,2,0],
                   'age':[80, 38, 12],
                   'height': [161.0, 173.5, 180.5],
                   'weight': [62.3, 55.7, 80.0]
                   })
df
```

```python
#Applicable to Pandas Series only
#Element-wise operation
#Primarily used for replacing values
#The arg parameter accepts a mapping between old value and new value can be in a form of (a) Dictionary, (b) Series and (c) Function
```

```python
gender_map = {0: 'Unknown', 1:'Male', 2:'Female'}
df['gender'] = df['gender'].map(gender_map)
df
```

```python
gender_map = {0: 'Unknown', 1:'Male', 2:'Female'}
s = pd.Series(gender_map) # mapping series
df['gender'] = df['gender'].map(s)
df
```

```python
df['age_group'] = df['age'].map(lambda x: 'Adult' if x >= 21 else 'Child')
df
```

**What is Pandas apply()?
.apply() is applicable to both Pandas DataFrame and Series. When applied to DataFrames, .apply() can operate row or column wise.**

```python
#Applicable to Pandas Series
#Accepts a function
#Able to pass positional or keyword arguments to function
#Can return either a Series or DataFrame
```

```python
def get_age_group(age, threshold):
    if age >= int(threshold):
        age_group = 'Adult'
    else:
        age_group = 'Child'
    return age_group
```

```python
# keyword argument
df['age_group'] = df['age'].apply(get_age_group, threshold = 21)
df
```

```python
def get_age_group(age, lower_threshold, upper_threshold):
    if age >= int(upper_threshold):
        age_group = 'Senior'
    elif age <= int(lower_threshold):
        age_group = 'Child'
    else:
        age_group = 'Adult'
    return age_group
```

```python
df['age_group'] = df['age'].apply(get_age_group, lower_threshold = 20, upper_threshold = 65)
df
```

```python
#Function can be applied either column-wise (axis = 0) or row-wise (axis = 1)
#Able to pass data as Series or numpy array to function
#Able to pass positional or keyword arguments to function

#we saw that .map() does not allow arguments to be passed into the function. .apply() on the other hand allows passing of both positional or keyword arguments..
# Lets parameterise the function to accept a thershold parameter.
```

```python
def get_last_name(x):
    return x.split(' ')[-1]
type(df['name'].apply(get_last_name))
df
#axis: Axis along which the function is applied. axis=0 - apply function to each column. axis=1 - apply function to each row.
```

```python
def some_math(x, multiplier, add):
    return x * multiplier + add
df = pd.DataFrame({'A':[1,2,3], 'B':[10,20,30]})
df.applymap(some_math, multiplier = 2, add = 1)
```
