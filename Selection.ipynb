```python
import pandas as pd
df=pd.read_csv("User.csv")
df.head()
```

```python
#Dealing with columns
#Column selection
df[['Age','User ID']].head(2)
```

```python
#Column addition
data={'Name':['suraj','Neeraj','Vipin'],
      'Roll_no':[1,2,3]}
df1=pd.DataFrame(data,index=['Rank1','Rank2','Rank3'])
df1
address=['PNP','Delhi','Uttrakhand']
df1['Address']=address #column addition
df1
```

```python
#Deletion Column
df1.drop(['Name','Roll_no'],axis=1)
df1
```

#Dealing with rows

**Pandas provide a unique method to retrieve rows from a dataframe using .loc()**

```python
#Using loc metod
df.loc[0:3,['Age','Gender']]
```

```python
df.columns
```

```python
#using range method
df.loc[0:3,'Age':'Purchased']
```

```python
#using boolean method
df.loc[df.Age>19,['Age','Gender']].head(3)
```

```python
#Boolean indexing we will select subsets of data based on the actual values of the data in the dataframe
#and not on their rows/columns or integer locatio.In boolean indexing we use a boolean vector to filter the data
df[df['Age']>19][['Age','Gender']].head(3)
```

```python
df[df['Age']==19].head()
```

```python
df[df.index==0]
```

```python
#Row addition
data={'Name':['suraj','Neeraj','Vipin'],
      'Roll_no':[1,2,3]}
df1=pd.DataFrame(data,index=['Rank1','Rank2','Rank3'])
df1
address=['PNP','Delhi','Uttrakhand']
df1['Address']=address #column addition
df1

new_row=pd.DataFrame({'Name':'Amit','Roll_no':4,'Address':'Chandigarh'},index=[4])
df2=pd.concat([new_row,df1]).reset_index(drop=True)
df2
```

```python
df.loc[[1,3],['Age','Gender']]
```
