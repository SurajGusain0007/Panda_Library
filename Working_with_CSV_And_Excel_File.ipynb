#Loading/imporing CSV Files

**read_csv is an important pandas function to read csv files and do operations on it**

```python
import pandas as pd
df=pd.read_csv('cardataset.csv')
df.head(1)
```

```python
# writing csv files with pandas
import pandas as pd
df.to_csv('suraj.csv')
df.head(1)
```

```python
#using index_col mak the passed column as index instead of 0,1,2,3,4in read_csv
#squeeze function is used to convert a single column dataframe into series
import pandas as pd
df=pd.read_csv('cardataset.csv',index_col='Make',squeeze=True)
df.head(2)
```

```python
#skips the passed rows in new series
import pandas as pd
df=pd.read_csv('cardataset.csv',skiprows=[1,2,3,4])
df.head(6)
```

#Loading Excel Spreadheet

```python
import pandas as pd
excel_file=pd.read_excel('User.xlsx')
excel_file.head()
```

```python
#Only read specific columns from an excel sheet
import pandas as pd
excel_file=pd.read_excel('User.xlsx',usecols=[0,3])
excel_file.head()
```
