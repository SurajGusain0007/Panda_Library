#Creating Objects

**Pandas Dataframe is two dimensional size mutable ,potentially heterogenous tabular data structure with lablled axes rows and columna i.e data is aligned in a tabular fashion in rows and columns**

```python
#Create a dataframe using list
import pandas as pd
list=['suraj','Neeraj','Ritu','Nisha','Vipin']
df=pd.DataFrame(list)
df
```

```python
#Create a dataframe using dictionary
df=pd.DataFrame({'Name':['suraj','Neeraj','Ritu','Nisha','Vipin'],
                 'Rollno':[1,2,3,4,5,],
                 'City':['Pnp','Kuk','Pnp','Nanital','Mat']})
df
```

```python
#Creating an Empty Dataframe
df=pd.DataFrame()
df
```

```python
#Creating dataframe from list of list
lis=[['suraj','pnp'],['Neeraj','kuk'],['Ritu','pnp'],['Nisha','Vipin']]
#lis1=['Pnp','Kuk','Pnp','Nanital','Mat']
df=pd.DataFrame(lis,columns=['Name','city'],index=[1,2,3,4])
df
                
```

```python
#Creating pandas dataframe list of dicts
data=[{'a':1,'b':2,'c':3},
      {'a':10,'b':20,'c':30}]
df=pd.DataFrame(data,index=['Rank1','Rank2'])
df      
```

```python
#increment the index so that index starts with 1(starts 0 by default)
data_rankings = {'test': ['India', 'South Africa', 'England', 'New Zealand', 'Australia'], 

              'odi': ['England', 'India', 'New Zealand', 'South Africa', 'Pakistan'], 

               't20': ['Pakistan', 'India', 'Australia', 'England', 'New Zealand']}
df=pd.DataFrame(data_rankings)   
df.index+=1 
df        
```

**Pandas Series is a one-dimensional labelled array capabale of holding any datatype integers ,float strings objects.The ases labels are collectively called index.Pandas series is nothing but column in a excel sheet**

```python
lis=[1,2,3,4,5]
df=pd.Series(lis)
df
```

```python
#Creating Series
import numpy as np
df=np.array(['a','b','c','d','e'])
df1=pd.Series(df)
df1
```

```python
#Creating Series using index
import numpy as np
df=np.array(['a','b','c','d','e'])
df1=pd.Series(df,index=[11,12,13,14,15])
df1
```
